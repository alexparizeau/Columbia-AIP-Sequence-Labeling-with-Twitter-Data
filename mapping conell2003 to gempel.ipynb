```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file = "oct27.supertsv"
data = open(file, "r",encoding='utf-8')
text = data.read()
print(text)
print(type(text))
```

```python
texts = text.split('\n')
```

```python
texts
```

```python
tagged_tweets = list()
for i in texts:
    if i.startswith('TWEET\t'):
        tweet = list()
    if not i.startswith('TWEET\t') and i!='TOKENS' and i != '':
        token = i[2:]
        tag = i[0]
        tweet.append((token,tag))
    if i == '':
        tagged_tweets.append(tweet)
```

```python
tagged_tweets
```

```python
#tag lists
tags = []
for t in tagged_tweets:
    #print([x[1] for x in t])
    tags.extend([x[1] for x in t])
len(tags)
```
